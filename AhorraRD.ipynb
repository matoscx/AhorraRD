## 3) Lista de compras

```python
shopping = pd.DataFrame([
    {'categoria':'Cereales','producto':'Arroz','unidad':'Libra','cantidad':2},
    {'categoria':'Legumbres','producto':'Habichuela','unidad':'Libra','cantidad':1},
    {'categoria':'Carnes','producto':'Pollo','unidad':'Libra','cantidad':1},
    {'categoria':'Aceite','producto':'Aceite Pequeno','unidad':'Unidad','cantidad':1},
    {'categoria':'Lacteos','producto':'Leche','unidad':'Unidad','cantidad':2},
    {'categoria':'Pan','producto':'Pan','unidad':'Unidad','cantidad':6},
    {'categoria':'Condimentos','producto':'Sal','unidad':'Libra','cantidad':1},
    {'categoria':'Huevos','producto':'Huevo','unidad':'1','cantidad':12},
    {'categoria':'Bebida','producto':'Cafe','unidad':'Unidad','cantidad':1},
    {'categoria':'Dulce','producto':'Pacofis','unidad':'Unidad','cantidad':1}
])
shopping
```

## 4) Cálculo

```python
from difflib import get_close_matches

def find_cheapest(prices_df, barrio, producto, unidad):
    df = prices_df[prices_df['barrio'].str.lower()==str(barrio).lower()]

    mask = (df['producto'].str.lower()==producto.lower()) & (df['unidad'].str.lower()==unidad.lower())
    if mask.any():
        cand = df[mask]
    else:

        cand = df[df['producto'].str.lower().str.contains(producto.lower(), na=False)]
    if cand.empty:
        return None
    best = cand.loc[cand['precio'].idxmin()]
    return {'colmado':best['colmado'],'precio':float(best['precio']),'producto':best['producto'],'unidad':best['unidad']}

def avg_price(prices_df, barrio, producto, unidad):
    df = prices_df[prices_df['barrio'].str.lower()==str(barrio).lower()]
    sub = df[(df['producto'].str.lower()==producto.lower()) & (df['unidad'].str.lower()==unidad.lower())]
    if sub.empty:
        sub = df[df['producto'].str.lower().str.contains(producto.lower(), na=False)]
    if sub.empty:
        return None
    return float(sub['precio'].mean())

def suggest_substitute(prices_df, barrio, producto, unidad, categoria=None):
    df = prices_df[prices_df['barrio'].str.lower()==str(barrio).lower()]

    candidates = df['producto'].dropna().unique().tolist()
    match = get_close_matches(producto, candidates, n=1, cutoff=0.5)
    if match:
        prod_name = match[0]
        dfm = df[df['producto']==prod_name]
        best = dfm.loc[dfm['precio'].idxmin()]
        explanation = f"Se sugiere {best['producto']} porque es similar a {producto} y es más barato en {best['colmado']}."
        return {'producto':best['producto'],'unidad':best['unidad'],'precio':float(best['precio']),'colmado':best['colmado'],'explanation':explanation}
    return None
```

## 5) IA Básica


```python
barrio_objetivo = combined['barrio'].mode().iloc[0]
print('Barrio objetivo:', barrio_objetivo)
results = []
total_opt = 0.0
total_avg = 0.0
for _, row in shopping.iterrows():
    prod = str(row['producto'])
    uni = str(row['unidad'])
    cant = float(row['cantidad'])
    best = find_cheapest(combined, barrio_objetivo, prod, uni)
    avg = avg_price(combined, barrio_objetivo, prod, uni)
    if best is not None:
        costo = best['precio'] * cant
        ahorro_item = (avg - best['precio']) * cant if avg is not None else 0.0
        results.append({'barrio':barrio_objetivo,'producto':prod,'unidad':uni,'cantidad':cant,'colmado_recomendado':best['colmado'],'precio_unitario_recomendado':best['precio'],'precio_promedio_barrio':round(avg,2) if avg is not None else None,'ahorro_vs_promedio':round(ahorro_item,2),'costo_total_item':round(costo,2),'sustituto_sugerido':'','explicacion_sustituto':''})
        total_opt += costo
        total_avg += (avg * cant) if avg is not None else best['precio'] * cant
    else:
        sub = suggest_substitute(combined, barrio_objetivo, prod, uni, row.get('categoria',None))
        if sub is not None:
            costo = sub['precio'] * cant
            results.append({'barrio':barrio_objetivo,'producto':prod,'unidad':uni,'cantidad':cant,'colmado_recomendado':sub['colmado'],'precio_unitario_recomendado':sub['precio'],'precio_promedio_barrio':None,'ahorro_vs_promedio':0.0,'costo_total_item':round(costo,2),'sustituto_sugerido':sub['producto'],'explicacion_sustituto':sub['explanation']})
            total_opt += costo
        else:
            results.append({'barrio':barrio_objetivo,'producto':prod,'unidad':uni,'cantidad':cant,'colmado_recomendado':None,'precio_unitario_recomendado':None,'precio_promedio_barrio':None,'ahorro_vs_promedio':0.0,'costo_total_item':0.0,'sustituto_sugerido':'','explicacion_sustituto':'No se encontró producto ni sustituto.'})

res_df = pd.DataFrame(results)
ahorro_total = round(total_avg - total_opt,2)
print('Costo total optimizado: RD$', round(total_opt,2))
print('Costo total si compra al precio promedio del barrio: RD$', round(total_avg,2))
print('Ahorro estimado: RD$', ahorro_total)
res_df
```
